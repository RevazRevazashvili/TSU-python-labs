# ფაილებთან მუშაობის მეთოდები

write()

```python
#  write() - ფუნქცია ახდენს პარამეტრად გადაცემული სტრიქონის ან ბაიტების მიმდევრობის ჩაწერას ფაილში. ფუნქცია აბრუნებს ჩაწერილი ბაიტების რაოდენობას.

f = open(file='data_text.txt', mode='wt', encoding='utf-8')
y = f.write('ეს სტრიქონი ჩაიწერება ფაილში');
print('Count of recorded bytes is ', y)
```

```python
f = open(file='data_binary.txt', mode='wb')
x = ('ეს სტრიქონი ჩაიწერება ფაილში').encode('utf-8')
y = f.write(x);
print('Count of recorded bytes is ', y)
```

writelines()

```python
# writelines() - ფუნქცია ახდენს პარამეტრად გადაცემული მიმდევრობის ჩაწერას ფაილში. ფუნქცია არაფერს არ აბრუნებს.

f = open(file='data_text.txt', mode='w', encoding='utf-8')
lst = ['ეს ', 'სტრიქონი ', 'ჩაიწერება ', 'ფაილში']
y = f.writelines(lst)
```

```python
f = open(file='data_binary.txt', mode='wb')
# მოვახდინოთ სიის ელემენტების კოდირება ‘utf-8’ ფორმატში
lst = ['ეს '.encode('utf-8'),'სტრიქონი '.encode('utf-8'),
'ჩაიწერება '.encode('utf-8'),'ფაილში'.encode('utf-8')]
y = f.writelines(lst);
```

writable()

```python
# writable() - ფუნქცია ამოწმებს ფაილი არის თუ არა გახსნილი ჩაწერის რეჟიმში. დადებითი პასუხის შემთხვევაში აბრუნებს True-ს, ხოლო წინააღმდეგ შემთხვევაში კი - False-ს.

f = open(file='data_text.txt', mode='r', encoding='utf-8')
# გვაინტერესებს ფაილი არის თუ არა გახსნილი ჩაწერის რეჟიმში
print('f1 file is open in write mode: ', f.writable())
```

read()

```python
# read() - ფუნქცია ახდენს პარამეტრად გადაცემული ბაიტების კითხვას ფაილიდან. თუ პარამეტრი არაა გადაცემული, მაშინ ხდება მონაცემების კითხვა მიმდინარე პოზიციიდან ფაილის ბოლომდე.

f = open(file='data_text.txt', mode='r', encoding='utf-8')
print('read file fully: ', f.read()) # მთლიანად წავიკითხოთ ფაილი
f.close() # დავხუროთ ფაილი
f = open(file='data_text.txt', mode='r', encoding='utf-8')
print('read only 7 byte', f.read(7)); # წავიკითხოთ მხოლდ 7 ბაიტი
```

readline()

```python
# readline() - ფუნქცია კითხულობს ფაილში ჩაწერილ მონაცემებიდან მხოლოდ ერთსტრიქონს ყოველ ჯერზე. ფაილის ბოლოში მისვლის შემთხვევაში აბრუნებს ცარიელ სტრიქონს.

f = open(file='data_text.txt', mode='r', encoding='utf-8')
# ფაილიდან წავიკითხოთ მონაცემები სტრიქონ-სტრიქონ
print('first: using readline : ', f.readline());
print('second: using readline : ', f.readline());
print('third; using readline : ', f.readline());
print('fourth: using readline : ', f.readline())
```

readlines()

```python
# readlines() - ფუნქცია კითხულობს ფაილში ჩაწერილ მთლიან ინფორმაციას და ინახავს მათ ცალცალკე სტრიქონად.

# ფაილთან მუშაობის ტექსტური რეჟიმი. გავხსნათ ფაილი
f = open(file='data_text.txt', mode='r', encoding='utf-8')
lst = f.readlines() # ფაილიდან ერთიანად წავიკითხოთ ყველა მონაცემი
i = 1 # მთვლელი სტრიქონის მნიშვნელობის მისათითებლად
# მოვახდინოთ ფაილში განთავსებული მონაცემების სტრიქონ-სტრიქონ ბეჭდვა
for st in lst:
 print('text in line', i , ': ', st)
 i += 1; # გავზარდოთ მთვლელის მნიშვნელობა
```

close()

```python
# close() - ფუნქცია ხურავს ფაილს. ფაილთან მუშაობის დასრულების შემდეგ საჭიროა ფაილის დახურვა. ფაილის დახურვის ოპერაცია გულისხმობს კავშირის გაწყვეტას შესაბამის ფაილთან.
```

```python
"""
Python ენაში მხარდაჭერილია კონტენტის მენეჯერი, რომელიც გარანტირებს ობიექტთან
მუშაობის დასრულებას მიუხედავად იმისა ობიექტის გამოყენებისას დაფიქსირდა თუ არა
შეცდომა. ამ მიზნით გამოიყენება ინსტრუქცია with as.
ითვლება, რომ with as ინსტრუქციის გამოყენებით ფაილის გახსნა უფრო უსაფრთხოა ვიდრე
ჩვეულებრივი მოქმედებით გახსნა და ამიტომ ფაილთან მუშაობის აუცილებლობის შემთხვევაში
რეკომენდირებულია ამ ინსტრუქციის გამოყენება.

"""
```
