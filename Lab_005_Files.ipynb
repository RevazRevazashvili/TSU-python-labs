writelines()

```python
# writelines() - ფუნქცია ახდენს პარამეტრად გადაცემული მიმდევრობის ჩაწერას ფაილში. ფუნქცია არაფერს არ აბრუნებს.

f = open(file='data_text.txt', mode='w', encoding='utf-8')
lst = ['ეს ', 'სტრიქონი ', 'ჩაიწერება ', 'ფაილში']
lst1 = ['ეს ', 'სტრიქონი ', 'ჩაიწერება ', 'ფაილში']
f.writelines(lst)
f.writelines(lst1)
f.close()

f = open(file='data_text.txt', mode='r', encoding='utf-8')
s = f.read()
f.close()
s
```

```python
f = open(file='data_binary.txt', mode='wb')
# მოვახდინოთ სიის ელემენტების კოდირება ‘utf-8’ ფორმატში
lst = ['ეს '.encode('utf-8'),'სტრიქონი '.encode('utf-8'),
'ჩაიწერება '.encode('utf-8'),'ფაილში'.encode('utf-8')]
y = f.writelines(lst);
```

```python
# read() - ფუნქცია ახდენს პარამეტრად გადაცემული ბაიტების კითხვას ფაილიდან. თუ პარამეტრი არაა გადაცემული, მაშინ ხდება მონაცემების კითხვა მიმდინარე პოზიციიდან ფაილის ბოლომდე.

f = open(file='data_text.txt', mode='r', encoding='utf-8')
print('read file fully: ', f.read()) # მთლიანად წავიკითხოთ ფაილი
f.close() # დავხუროთ ფაილი
f = open(file='data_text.txt', mode='rb')
st = f.read(7)
print('read only 7 byte', st.decode()); # წავიკითხოთ მხოლდ 7 ბაიტი
```

ფაილური ობიექტები StringIO და BytesIO

```python
"""
Python ენის io მოდულში არსებული კლასები StringIO და BytesIO იძლევა ოპერატიულ
მეხსიერებაში განთავსებულ სტრიქონთან და მონაცემთა ბაიტებთან მუშაობის საშუალებას
როგორც შესაბამისი ტიპის ობიექტებთან. ოპერატიულ მეხსიერებაში განთავსებული მონაცემების
მისაღებად გამოიყენება ფუნქცია getvalue().
შევნიშნოთ, რომ ფაილებთან მუშაობისთვის განკუთვნილი შესაბამისი მეთოდებიდან
უმეტესი შეიძლება გამოყენებული იქნას ამ შემთხვევაშიც.

"""
```

```python
import io # კოდის ფაილთან მივაერთოთ io მოდული

# სტრიქონულ ობიექტში ჩავწეროთ სტრიქონი
x = 'ეს სტრიქონი ჩაიწერება ოპერატიულ მეხსიერებაში.'

f1 = io.StringIO(x) # შევქმნათ სტრიქონული ობიექტი

f1.write(x) # ჩავწეროთ ოპერატიული მეხსიერებიდან x ცვლადის შიგთავსი
print(f1.getvalue()) # დავბეჭდოთ სტრიქონული ობიექტის შიგთავსი
f1.close() # გავწყვიტოთ კავშირი სტრიქონულ ობიექტთან

print('f2: read data from memory:')
x = 'ეს სტრიქონი იქნება წაკითხული ოპერატიულ მეხსიერებიდან.'
f2 = io.StringIO(x) # შევქმნათ სტრიქონული ობიექტი

# სტრიქონული ობიექტიდან წავიკითხოთ ტექსტური მონაცემი და დავბეჭდოთ ეს მონაცემი
print(f2.read())
f2.close() # გავწყვიტოთ კავშირი სტრიქონულ ობიექტთან
```

```python
import io

x = 'ეს სტრიქონი ჩაიწერება ოპერატიულ მეხსიერებაში.'.encode()

f1 = io.BytesIO(x)

f1.write(x) # ჩავწეროთ ოპერატიული მეხსიერებიდან x ცვლადის შიგთავსი
print(f1.getvalue()) # დავბეჭდოთ სტრიქონული ობიექტის შიგთავსი
f1.close() # გავწყვიტოთ კავშირი სტრიქონულ ობიექტთან

print('f2: read data from memory:')
x = 'ეს სტრიქონი იქნება წაკითხული ოპერატიულ მეხსიერებიდან.'.encode()
f2 = io.BytesIO(x) # შევქმნათ სტრიქონული ობიექტი

# სტრიქონული ობიექტიდან წავიკითხოთ ტექსტური მონაცემი და დავბეჭდოთ ეს მონაცემი
print(f2.read().decode())
f2.close() # გავწყვიტოთ კავშირი სტრიქონულ ობიექტთან
```

```python
def filler1(file, a, b):
  with open(file, 'a') as f:
    for _ in range(10):
      f.write(str(random.randint(a, b)) + " ")
```

```python

def read_files(file, c, d):
  with open(file, 'r') as f:
    lst: list[str] = f.readlines()
  dict_counter: dict[int, int] = {}
  def count():
    for i in range(len(lst)):
      counter: int = 0
      for word in lst[i].split():
        if c <= int(word) <= d:
          counter += 1
          dict_counter[str(i)] = counter
  count()
  return dict_counter
```

```python
def write_lines(file: str) -> None:
  with open(file, 'r') as f:
    numbers: str = f.readlines()[0].split()
  for i in range(1000, 10000, 1000):
    for number in numbers:
      if i <= int(number) <= i + 999:
        with open(f"{i}.in", "a") as f:
          f.write(number + " ")

write_lines("data.txt")
```

```python
def count() -> list[int]:
  lst: list[int] = []
  for i in range(1000, 10000, 1000):
    with open(f"{i}.in", "r") as f:
      numbers: list[str] = f.readlines()[0].split()
      lst.append(len(numbers))
  return lst

count()
```

```python
# გვარების სიმრავლე: 100 ჩანაწერი
LNames = ['აბაშიძე', 'გიგაური', 'არჩვაძე', 'ახალაია', 'ბაძაღუა', 'ბერიანიძე',
'ბერიშვილი', 'გვენცაძე', 'დალაქიშვილი', 'ანთიძე', 'გიორგაძე', 'გოგალაძე', 'გოცირიძე',
'ვარდიძე', 'ზარანდია', 'თადუმაძე', 'ლაბაძე', 'კვარაცხელია', 'კუსრაძე', 'კვესელავა',
'კაპანაძე', 'კასრაძე', 'კვინიკაძე']
# print(len(LNames))
# გოგონების სახელების სიმრავლე: 40 ჩანაწერი
GNames = ['ანა', 'ანუკი', 'ბარბარე', 'გვანცა', 'დიანა', 'ეკა', 'ელენე', 'ვერონიკა',
'ვიქტორია', 'თათია', 'ლამზირა']
# print(len(GNames))
# ბიჭების სახელების სიმრავლე: 40 ჩანაწერი
BNames = ['ალეკო', 'ალიკა', 'ამირან', 'ანდრია', 'არჩილი', 'ასლანი', 'ბაჩუკი', 'ბექა',
'გიგა', 'გიორგი', 'დავითი', 'გიგი', 'გოგა', 'დათა', 'ერეკლე', 'თემური', 'იაკობ',
'ილია', 'ირაკლი', 'ლადო']
# print(len(BNames))
# ერთად აღებული გოგონების და ბიჭების სახელების სიმრავლე: 80 ჩანაწერი
FNames = GNames + BNames
# print(len(FNames))
# საგნების დასახელებების სიმრავლე: 10 ჩანაწერი
Subject = ['პროგრამირების საფუძვლები', 'კალკულუსი II', 'შესავალი ფიზიკაში',
           'კომპიუტერული უნარჩვევები', 'მონაცემთა სტრუქტურები', 'ალგორითმები II']
""" შენიშვნა. ქვემოთ მოყვანილი კოდი Stud_list სიმრავლეს ავსებს 100000 ჩანაწერით """
Point = [str(i) for i in range(101)] # შეფასების სიმრავლე
from random import choice as ch
# სტუდენტებზე ჩანაწერების სიმრავლე
Stud_recs = [' '.join([ch(LNames), ch(FNames), ch(Subject), ch(Point)]) for _ in
range(100)]

```

```python
print(Stud_recs)
```

```python
def write_file():
  for rec in Stud_recs:
    sname = rec.split()[0]
    with open(f"{sname}.txt", "a") as f:
      f.write(rec + "\n")

write_file()
```

```python
def max_number() -> str:
  for rec in Stud_recs:
    sname = rec.split()[0]
    max = -1
    with open(f"{sname}.txt", "r") as f:
      students = f.readlines()
      file_name = ""
      if len(students) > max:
        max = len(students)
        file_name = sname
  return file_name, max
max_number()
```
